```python
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
file_path = 'data/winequality-white.csv'
df = pd.read_csv(file_path, sep=";")
df.head()
```

```python
# check for nulls ( no nulls), and data type
df.info()
```

```python
# separate features from target
y=df['quality']
X=df.drop(columns='quality')
```

```python
# clean it up, scale it etc
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
X_scaled_df.hist(bins=15, figsize=(15, 10))
plt.suptitle('Scaled Feature Distributions', fontsize=16)
plt.show()
```

```python
import seaborn as sns
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

for feature in features:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=feature, y='quality', data=df)
    plt.title(f'{feature} vs. Wine Quality')
    plt.xlabel(feature)
    plt.ylabel('Quality')
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='quality', data=df)  # Assuming 'df' is your DataFrame
plt.title('Distribution of Wine Quality Ratings')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Wine Features')
plt.show()
```

```python
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='quality', y=feature, data=df)
    plt.title(f'{feature} Distribution by Wine Quality')
    plt.xlabel('Quality')
    plt.ylabel(feature)
    plt.show()
```

```python
# train test module
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1776, 
                                                    stratify=y)
X_train.shape
```

```python
# logistic regress
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                               max_iter=200,
                               random_state=1776)
classifier.fit(X_train, y_train)
```

```python
# predictions and confusion matrix
testing_predictions = classifier.predict(X_test)
test_matrix = confusion_matrix(y_test, testing_predictions)
print(test_matrix)
```

```python
# classification report
test_report = classification_report(y_test, testing_predictions)
print(test_report)
```
